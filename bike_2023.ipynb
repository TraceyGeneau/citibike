```python
#import dependancies
import pandas as pd
pd.set_option('max_colwidth',400)
```

```python
#assign dtype to datafame https://bobbyhadz.com/blog/columns-have-mixed-types-specify-dtype-option-on-import
dtype={'start_station_id': object,
       'end_station_id':object
     }
```

```python
#combine list of files
files2023 = ['2023/JC-202303-citibike-tripdata.csv','2023/JC-202304-citibike-tripdata.csv','2023/JC-202305-citibike-tripdata.csv','2023/JC-202306-citibike-tripdata.csv','2023/JC-202307-citibike-tripdata.csv']

```

```python
# Create an empty list to store each dataframe
bike_2023_list = []
```

```python
# Loop through each file in the list
for file in files2023:
   
    # Read the file into a dataframe
    bike_2023 = pd.read_csv(file, sep=',',
    encoding='utf-8',
    dtype=dtype)

    # Append the dataframe to the list
    bike_2023_list.append(bike_2023)
```

```python
# Concatenate all dataframes in the list into one dataframe
bike_2023 = pd.concat(bike_2023_list)
```
